## Descenso del gradiente

## Dependencias

```python
from matplotlib import cm # Para manejar colores
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import random
```

### Gr치fica en 3D de nuestra funci칩n de coste

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

def f(x,y):
  return x**2 + y**2;

resolution = 100
X = np.linspace(-4, 4, resolution)
Y = np.linspace(-4, 4, resolution)
X, Y = np.meshgrid(X, Y)

Z = f(X,Y) 

# Gr치ficar la superficie
surf = ax.plot_surface(X, Y, Z, cmap=cm.cool,
                       linewidth=0, antialiased=False)
fig.colorbar(surf)
```

## Descenso del gradiente en una funci칩n

```python
figure(figsize=(9, 8), dpi=80)

# Crear plano cartesiano
resolution = 100
X = np.linspace(-80, 100, resolution)
Y = np.linspace(-100, 100, resolution)
X, Y = np.meshgrid(X, Y)

# Funcion de interes
def f(x,y):
  return (x-2)**2 + 3*(y+1)**2;
Z = f(X,Y) 

# Curvas de nivel 
plt.contour(X, Y, Z, levels=10, cmap='rainbow')
plt.colorbar()
plt.title('Descenso del gradiente')

# Derivada de la funcion utilizando definicion por limite
h = 0.01
def derivate(_p,p):
    return  (f(_p[0],_p[1]) - f(p[0],p[1])) / h

# Seleccion aleatoria del punto de partida
p = np.random.rand(2) * 95 
plt.plot(p[0],p[1],'o', c='r')


alpha = 0.01
grad = np.zeros(2)

for i in range(10000):
    for idx, val in enumerate(p):
        
        _p = np.copy(p)
        _p[idx] = _p[idx] + h;
        dp = derivate(_p,p) 
        grad[idx] = dp

    p = p - alpha * grad

    if(i % 10 == 0):
        plt.plot(p[0],p[1],'o', c='c')

plt.plot(p[0],p[1],'o', c='m')
plt.show()

print("El punto m칤nimo se encuentra en: ", p)
```

## Descenso Gradiente en optimizaci칩n de parametros

```python
resolution = 100
x = np.linspace(-8.0, 8.0, resolution)

## Crear la funcion base y agregar ruido a esta para efectos
## generar un ejemplo cercano a la realidad
y = 10 + 2*np.power(x,2)  # a + bx^2 <> 10 + 2x^2
y_noise = 5 * np.random.normal(size=x.size)
ydata = y + y_noise

# Seleccion aleatoria de los parametros a y b de la funcion a + bx^2
parameter[0] = random.uniform(-1, 1) * 145
parameter[1] = random.uniform(-1, 1) * 75

print('Paramtros generados aleatoriamente') 
print('a:', parameter[0], ', b:', parameter[1]) 

# Funcion que puede describir el comportamiento del conjunto de datos
def y_predict(x,a,b):
  return  a + np.power(x,2)*b

# Graficar datos y funciones
y_prediction = y_predict(x,parameter[0],parameter[1])
figure(figsize=(8, 7), dpi=80)
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'w')
plt.plot(x,y_prediction, 'r')
plt.title('Estado inicial, con parametros seleccionados aleatoriamente')
plt.ylabel(' f(x) = y = a + bx^2')
plt.xlabel('x')
plt.show()


# Crear plano cartesiano
figure(figsize=(14, 12), dpi=80)
A = np.linspace(-150, 150, resolution)
B = np.linspace(-80, 80, resolution)
A, B = np.meshgrid(A, B)

# SEE : Para generar un mapa en curvas de nivel de a, b.
def see_(y,a,b):
    return np.power(y - y_predict(x, a, b), 2)
SEE = see_(ydata,A,B)

# Curvas de nivel 
plt.contour(A, B, SEE, levels=100, cmap='rainbow')
plt.colorbar()
plt.title('Descenso del gradiente')
plt.ylabel('parametro b')
plt.xlabel('parametro a')


# MSE : Funcion que se va a optimizar
def mse(y,a,b):
    return sum(np.power(y - y_predict(x, a, b), 2)) * (1/len(y))

# Derivada de la funcion MSE usando la definicion por limite
h = 0.01
def derivate_mse(_parameter,parameter):
    return  (mse(ydata,_parameter[0],_parameter[1]) - mse(ydata,parameter[0],parameter[1])) / h

# Graficar punto inicial en rojo
plt.plot(parameter[0],parameter[1],'o', c='r')

#
def alpha(param,grad,alpha):
    return param - alpha * grad
h1 = 0.01
#h2 = 0.000005
def derivate_alpha(param,grad,_alpha,alpha):
    return  (param - _alpha * min(grad) - (param - alpha * min(grad))) / h

# Proceso iterativo usando metodo de gradientes
alpha = 0.001
grad = np.zeros(2)
for i in range(10000):
    for idx, val in enumerate(parameter):

        _parameter = np.copy(parameter)
        _parameter[idx] = _parameter[idx] + h1;
        dp = derivate_mse(_parameter,parameter)
        grad[idx] = dp
    
    #print(alpha)
    #_alpha = np.copy(alpha)
    #_alpha = alpha + h2;
    #db_alfa = derivate_alpha(parameter, grad, _alpha, alpha)
    #alpha = alpha * db_alfa
    
    parameter = parameter - alpha * grad

    if(i % 250 == 0):
        plt.plot(parameter[0],parameter[1],'o', c='c')

# Graficar punto final en morado        
plt.plot(parameter[0],parameter[1],'o', c='m')
plt.show()


mse = mse(x,parameter[0],parameter[1])
print('mse:', mse)
print('****************************************')
print('Parametros ajustados') 
print('a:', parameter[0], ', b:', parameter[1]) 

# funcion a + bx^2 con parametros ajustados
y_prediction = y_predict(x,parameter[0],parameter[1])

# Graficar comprobaci칩n
figure(figsize=(8, 7), dpi=80)
plt.plot(x, ydata,  'bo')
plt.plot(x,y_prediction, 'r')
plt.title('Estado final, con parametros ajustados')
plt.ylabel(' f(x) = y = a + bx^2')
plt.xlabel('x')
plt.show()
```
